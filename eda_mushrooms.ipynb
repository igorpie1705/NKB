### Informacje o typach danych
Sprawdzam, ile jest zmiennych kategorycznych w zbiorze danych, jakie są ich nazwy oraz czy występują brakujące wartości.

```python
categorical = [var for var in df.columns if df[var].dtype=='O']

len(categorical)
```

```python
categorical
```

```python
df[categorical].isnull().sum()
```

Czyli w zbiorze danych mamy 23 zmienne kategoryczne, a żadna z nich nie posiada brakujących wartości.

### Sprawdzenie unikalnych wartości w zmiennych kategorycznych
Przy okazji sprawdzam, czy w zbiorze danych występują jakieś wartości, które nie są liczbami, a znakami zapytania. Jeśli tak, to zamieniam je na wartości NaN.

```python
for var in categorical:
    print(df[var].value_counts())

print(df['stalk-root'].unique())

df['stalk-root'] = df['stalk-root'].replace('?', np.nan)
```

```python
for var in categorical:
    print(df[var].value_counts())
```

Jak widać, w stalk-root występowały wartości "?" które zostały wyeliminowane.
